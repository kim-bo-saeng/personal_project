```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
from itertools import combinations
pd.set_option('display.max_columns',200)
```

```python
data_meta_row = pd.read_csv('c:\\Users\\user\\Documents\\GitHub\\data_tb_file\\lol_metadata_tb.csv')
```

```python
player_df = data_meta_row[['matchId','player_1','player_2','player_3','player_4','player_5','player_6','player_7','player_8','player_9','player_10']]
player_df
```

```python
player_match_df = pd.DataFrame()

for i in range(200):
    combi_list = list(player_df.loc[i])[1:11]
    combi_sets = []
    for combi_set in combinations(combi_list,2):
        combi_sets += [combi_set]

    df = pd.DataFrame(combi_sets,columns=['player_A','player_B'])
    df['matchId'] = player_df.loc[i][0]
    df = df[['matchId','player_A','player_B']]

    player_match_df = pd.concat([player_match_df,df],ignore_index = True)

player_match_df.to_csv('c:\\Users\\user\\Documents\\GitHub\\data_tb_file\\player_match_tb_test2.csv')
```
