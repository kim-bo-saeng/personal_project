```python
import requests
import pandas as pd
import numpy as np
import time
```

```python
# 첼린저,그랜드마스터,마스터의 솔로랭크 정보를 가져오기 위한 함수 생성

import requests
 
api_key = "RGAPI-f9ad895b-5f33-4894-8069-db22118e56a0"
header = {"X-Riot-Token" : api_key}
 
def league_v4_tier(tier):
    # 솔랭 기준
    queue = 'RANKED_SOLO_5x5'
 
    if tier == 'challenger' :
        url = f"https://kr.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/{queue}"
    elif tier == 'grandmaster':
        url = f"https://kr.api.riotgames.com/lol/league/v4/grandmasterleagues/by-queue/{queue}"
    else:
        url = f"https://kr.api.riotgames.com/lol/league/v4/masterleagues/by-queue/{queue}"
        
    return requests.get(url, headers=header)
```

```python
# 티어별 솔로랭크 정보를 테이블로 만들기

challenger_data = pd.DataFrame(league_v4_tier('challenger').json())
grandmaster_data = pd.DataFrame(league_v4_tier('grandmaster').json())
master_data = pd.DataFrame(league_v4_tier('master').json())
```

```python
# 첼린저 티어의 솔로랭크 정보중 entries에 들어 있는 summonerId 및 승패정보를 테이블로 담기

challenger_df = pd.DataFrame([challenger_data['entries'][0]])
data_size = len(challenger_data)

for i in range(1,data_size):
    append_df = pd.DataFrame([challenger_data['entries'][i]])
    challenger_df = pd.concat([challenger_df,append_df],ignore_index = True)

challenger_df['tier'] = 'CHALLENGER'
challenger_df
```

```python
# 그랜드마스터 티어의 솔로랭크 정보중 entries에 들어 있는 summonerId 및 승패정보를 테이블로 담기

grandmaster_df = pd.DataFrame([grandmaster_data['entries'][0]])
data_size = len(grandmaster_data)

for i in range(1,data_size):
    append_df = pd.DataFrame([grandmaster_data['entries'][i]])
    grandmaster_df = pd.concat([grandmaster_df,append_df],ignore_index = True)

grandmaster_df['tier'] = 'GRANDMASTER'
grandmaster_df
```

```python
# 마스터 티어의 솔로랭크 정보중 entries에 들어 있는 summonerId 및 승패정보를 테이블로 담기

master_df = pd.DataFrame([master_data['entries'][0]])
data_size = len(master_data)

for i in range(1,data_size):
    append_df = pd.DataFrame([master_data['entries'][i]])
    master_df = pd.concat([master_df,append_df],ignore_index = True)

master_df['tier'] = 'MASTER'
master_df
```

```python
# 첼린저,그마,마스터 티어의 entries정보를 하나의 테이블로 합치기

top_class_df = pd.concat([challenger_df,grandmaster_df,master_df],ignore_index = True)
top_class_df
```
